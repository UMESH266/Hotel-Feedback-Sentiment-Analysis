#### Text preprocessing

```python
# Necessary libraries
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

# Word cloud visualization
from wordcloud import WordCloud

# Stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Text preprocessing function
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove non-alphabetic characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Tokenize the text
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.lower() not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Remove common nouns like "and", "is", "the", and specific nouns like "rooms", "hotels", etc.
    common_nouns = set(["and", "is", "the","near","also"])  
    tokens = [word for word in tokens if word.lower() not in common_nouns]
    
    # Filter out words with length <= 2 (adjust as needed)
    tokens = [word for word in tokens if len(word) > 2]
    
    # Join the tokens back into a string
    processed_text = ' '.join(tokens)
    
    return processed_text
```

```python
feedback['Processed FEED BACK'] = feedback['FEED BACK'].apply(preprocess_text)
feedback.to_csv('Feedback_std.csv')
```

#### Bigrams

```python
# Bi / Tri gram generator function
def ngrams(corpus, ngram, n=None):
    # ngram_range=(2, 2)
    vec1 = CountVectorizer(ngram_range=ngram, max_features=2000).fit(corpus) 
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec1.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
# Extraction of bigrams
top2_words = ngrams(feedback["Processed FEED BACK"], ngram=(2,2), n=20)
top2_df = pd.DataFrame(top2_words)
top2_df.columns = ["Bi-gram", "Freq"]
top2_df.head(10)
```

```python
adjective_counter = {}

# Iterate over tokens in the doc_block
for i in doc_block:
    if i.pos_ == "ADJ":  # Check if the token is an adjective
        adjective = i.text.lower()  # Convert the adjective to lowercase
        if adjective in adjective_counter:
            adjective_counter[adjective] += 1  # Increment the count of the adjective
        else:
            adjective_counter[adjective] = 1  # Initialize the count for the adjective

sorted_adjectives = sorted(adjective_counter.items(), key=lambda x: x[1], reverse=True)
adjectives = [item[0] for item in sorted_adjectives]
counts = [item[1] for item in sorted_adjectives]

# Plot the graph
plt.figure(figsize=(10, 6))
plt.bar(adjectives, counts)
plt.xlabel("Adjectives")
plt.ylabel("Frequency")
plt.title("Frequency of Adjectives in Text Block")
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

#### BERT

```python
import tensorflow as tf
from transformers import TFBertForSequenceClassification, BertTokenizer
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
# Pretrained model for BERT
MODEL = "bert-base-uncased"
tokenizer_bert = AutoTokenizer.from_pretrained(MODEL)
model_bert = AutoModelForSequenceClassification.from_pretrained(MODEL)

def bert_polarity_score(text):
    encoded_text = tokenizer_bert(text, return_tensors='pt')
    output = model_bert(**encoded_text)
    
    # Get the number of output classes from the model's configuration
    num_classes = model_bert.config.num_labels
    
    scores = output.logits[0].detach().numpy()
    scores = softmax(scores)
    
    # Depending on the number of output classes, adjust the scores dictionary
    if num_classes == 2:  # Binary classification (e.g., negative and positive)
        scores_dict = {
            'Bert_Negative': scores[0],
            'Bert_Positive': scores[1]
        }
    elif num_classes == 3:  # Multi-class classification (e.g., negative, neutral, and positive)
        scores_dict = {
            'Bert_neg': scores[0],
            'Bert_neu': scores[1],
            'Bert_pos': scores[2]
        }
    else:
        raise ValueError(f"Unsupported number of output classes: {num_classes}")
    
    return scores_dict

# Assuming 'feedback' is your DataFrame containing feedback data
bert_results = feedback['Processed FEED BACK'].apply(lambda x: bert_polarity_score(x))
bert_results
```

#### RoBERTa model

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
import pickle
```

```python
# Pretrained model building using twitter sentiments
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
with open('roberta.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
# Roberta model
def roberta_polarity_score(text):
    encoded_text = tokenizer(text, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg': scores[0],
        'roberta_neu': scores[1],
        'roberta_pos': scores[2],
    }

    return scores_dict
```
